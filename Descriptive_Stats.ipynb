# Descriptive Statistics

Se aplica estadística descriptiva sobre el dataset de vinos tintos para ver medidas de tendencia central y dispersión, junto con gráficas.

Importación y carga de datos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"
wine = pd.read_csv(url, delimiter=";")

wine.head()
```

Medidas de Tendencia Central

Media.

```python
means = wine.mean()
print("Medias:\n", means)

print(f"Mean alcohol: {wine['alcohol'].mean():.2f}")

means.plot(kind="bar", figsize=(8,4), color="skyblue", title="Mean values")
plt.show()
```

Mediana.

```python
medians = wine.median()
print("Medianas:\n", medians)
print(f"Median alcohol: {wine['alcohol'].median():.2f}")

medians.plot(kind="bar", figsize=(8,4), color="salmon", title="Median values")
plt.show()
```

Moda.

```python
modes = wine.mode().iloc[0]
print("Modas:\n", modes)
print(f"Mode alcohol: {wine['alcohol'].mode()[0]:.2f}")

modes.plot(kind="bar", figsize=(8,4), color="lightgreen", title="Mode values")
plt.show()
```

Distribución y dispersión

Histograma con moda de alcohol

```python
sns.histplot(wine["alcohol"], bins=30, color="skyblue")
mode_val = wine["alcohol"].mode()[0]
plt.axvline(mode_val, color="red", linestyle="--")
plt.title("Distribución de Alcohol")
plt.show()
```

Rango

```python
rango = wine["alcohol"].max() - wine["alcohol"].min()
print(f"Range alcohol: {rango:.2f}")
```

Varianza y Desviación Estándar

```python
var = wine["alcohol"].var()
std = wine["alcohol"].std()
print(f"Variance: {var:.2f}, Std Dev: {std:.2f}")

sns.histplot(wine["alcohol"], bins=30, kde=True, color="skyblue")
plt.axvline(wine["alcohol"].mean(), color="red", linestyle="--", label="Mean")
plt.axvline(wine["alcohol"].mean()+std, color="green", linestyle="--", label="+1 SD")
plt.axvline(wine["alcohol"].mean()-std, color="green", linestyle="--", label="-1 SD")
plt.legend()
plt.show()
```

Cuartiles e IQR

```python
Q1, Q2, Q3 = np.percentile(wine["alcohol"], [25,50,75])
IQR = Q3 - Q1
print(f"Q1={Q1:.2f}, Q2={Q2:.2f}, Q3={Q3:.2f}, IQR={IQR:.2f}")

sns.boxplot(x=wine["alcohol"], color="lightblue")
plt.title("Boxplot Alcohol")
plt.show()
```
