```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"
df = pd.read_csv(url, parse_dates=['Month'], index_col='Month')
df.head(10)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df, color='blue', marker='o', linestyle='-', linewidth=2)
plt.title('Monthly Passengers Count')
plt.xlabel('Year')
plt.ylabel('Passengers Count')
plt.grid(True)
plt.show()
```

```python
years = df.index.year.unique()
plt.figure(figsize=(12,6))
for year in years:
    year_data = df[df.index.year == year]
    plt.plot(year_data['Passengers'], label=year)

plt.title('Monthly Passengers Count by Year')
plt.xlabel('Month')
plt.ylabel('Passengers Count')
plt.legend()
plt.grid(True)
plt.show()
```

```python
df['Year'] = df.index.year
df['month'] = df.index.month
table = pd.pivot_table(df, values='Passengers', index=['month'], columns='Year', aggfunc='sum')
fig, ax = plt.subplots(figsize=(12, 6))
table.plot(ax=ax, kind='line', legend=True)
plt.title('Seasonal Subseries')
plt.xlabel('Month')
plt.ylabel('Passengers Count')
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(df['Passengers'], bins=20, kde=True)
plt.title('Histogram of Passengers Count')
plt.xlabel('Passengers Count')
plt.ylabel('Frequency')
plt.subplot(1, 2, 2)
sns.kdeplot(df['Passengers'], fill=True)
plt.title('Density Plot of Passengers Count')
plt.xlabel('Passengers Count')
plt.ylabel('Density')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df_sun = pd.read_csv("Sunspots.csv")
df_sun['Date'] = pd.to_datetime(df_sun['Date'])
df_sun = df_sun.set_index('Date')
df_sun.rename(columns={'Monthly Mean Total Sunspot Number': 'Sunspots'}, inplace=True)
print(df_sun.head())

plt.figure(figsize=(12, 6))
plt.plot(df_sun['Sunspots'], marker='o', linestyle='-', linewidth=1)
plt.title('Monthly Sunspot Count')
plt.xlabel('Year')
plt.ylabel('Sunspots')
plt.grid(True)
plt.show()

years = df_sun.index.year.unique()
plt.figure(figsize=(12,6))
for year in years[:20]:
    year_data = df_sun[df_sun.index.year == year]
    plt.plot(year_data['Sunspots'], label=year)
plt.title('Sunspots by Year')
plt.xlabel('Month')
plt.ylabel('Sunspots')
plt.legend()
plt.grid(True)
plt.show()

df_sun['Year'] = df_sun.index.year
df_sun['Month'] = df_sun.index.month

table_sun = pd.pivot_table(df_sun, values='Sunspots', index='Month', columns='Year', aggfunc='mean')
plt.figure(figsize=(12, 6))
table_sun.plot(figsize=(12,6))
plt.title('Seasonal Subseries of Sunspots')
plt.xlabel('Month')
plt.ylabel('Sunspots')
plt.show()
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.histplot(df_sun['Sunspots'], bins=20, kde=True)
plt.title('Histogram of Sunspots')

plt.subplot(1, 2, 2)
sns.kdeplot(df_sun['Sunspots'], fill=True)
plt.title('Density Plot of Sunspots')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df_unemp = pd.read_csv("Unemployment.csv")
df_unemp = df_unemp.rename(columns={
    "Percent (%) of Labor Force Unemployed in State/Area": "Unemployment_rate"
})
df_unemp['Date'] = pd.to_datetime(
    df_unemp['Year'].astype(str) + '-' + df_unemp['Month'].astype(str) + '-01'
)

df_unemp = df_unemp.set_index('Date')

print(df_unemp.head())
state = "California"
df_state = df_unemp[df_unemp['State/Area'] == state]

plt.figure(figsize=(12,6))
plt.plot(df_state['Unemployment_rate'], marker='o', linestyle='-', linewidth=1)
plt.title(f'Unemployment Rate in {state}')
plt.xlabel('Year')
plt.ylabel('Unemployment Rate')
plt.grid(True)
plt.show()
years = df_state.index.year.unique()
plt.figure(figsize=(12,6))
for year in years:
    year_data = df_state[df_state.index.year == year]
    plt.plot(year_data['Unemployment_rate'], label=str(year))
plt.title(f'Unemployment Rate by Year in {state}')
plt.xlabel('Month')
plt.ylabel('Rate')
plt.legend()
plt.grid(True)
plt.show()
df_state['Year'] = df_state.index.year
df_state['Month'] = df_state.index.month

table_unemp = pd.pivot_table(df_state, values='Unemployment_rate',
                             index='Month', columns='Year', aggfunc='mean')

plt.figure(figsize=(12,6))
table_unemp.plot(figsize=(12,6))
plt.title(f'Seasonal Subseries  Unemployment in {state}')
plt.xlabel('Month')
plt.ylabel('Rate')
plt.show()
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
sns.histplot(df_state['Unemployment_rate'], bins=20, kde=True)
plt.title(f'Histogram  {state}')

plt.subplot(1,2,2)
sns.kdeplot(df_state['Unemployment_rate'], fill=True)
plt.title(f'Density  {state}')
plt.show()
```
