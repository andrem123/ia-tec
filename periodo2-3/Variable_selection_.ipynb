```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

ruta_archivo = "calhousing.csv"
housing_df = pd.read_csv(ruta_archivo)
housing_df
```

```python
nulos_por_columna = housing_df.isnull().sum()
nulos_por_columna
```

```python
mediana_dormitorios = housing_df["total_bedrooms"].median()
housing_df["total_bedrooms"] = housing_df["total_bedrooms"].fillna(mediana_dormitorios)
housing_df["total_bedrooms"]
```

```python
resumen_housing = housing_df.describe(include="all").T
resumen_housing
```

```python
housing_df.info()
```

```python
housing_num = housing_df.drop(columns="ocean_proximity")
housing_num.info()
matriz_corr = housing_num.corr()
matriz_corr
fig, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(matriz_corr, annot=True, fmt=".2f", cmap="coolwarm")
ax.set_title("Matriz de correlaciones", fontsize=18)
ax.set_xlabel("Variables", fontsize=14)
ax.set_ylabel("Variables", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
corr_target = matriz_corr["median_house_value"].sort_values(ascending=False)
corr_target
fig, ax = plt.subplots(figsize=(8, 5))
sns.barplot(x=corr_target.index, y=corr_target.values, ax=ax)
ax.set_title("Correlación con median_house_value", fontsize=16)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
