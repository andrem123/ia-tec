```python
import seaborn as sns
import pandas  as pd
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
print(titanic.isnull().sum())
```

```python
data = titanic[['age', 'fare', 'pclass', 'sex']]
data = pd.get_dummies(data, drop_first=True)
data.head()
```

```python
imputer = IterativeImputer(max_iter=10, random_state=0)
data_imputed = pd.DataFrame(imputer.fit_transform(data), columns=data.columns)
```

```python
print("Datos originales:")
print(data.head(30))

print("\nDatos imputados:")
print(data_imputed.head(30))
```

```python
import pandas as pd
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

life = pd.read_csv("Life Expectancy.csv")
print(life.head())
print(life.isnull().sum())

life_numeric = life.select_dtypes(include=['float64', 'int64'])

imputer = IterativeImputer(max_iter=10, random_state=0)

life_imputed = pd.DataFrame(
    imputer.fit_transform(life_numeric),
    columns=life_numeric.columns
)

print("Original data with missing values")
print(life_numeric.head(20))
print("\nData after MICE imputation")
print(life_imputed.head(20))
```

```python
import seaborn as sns
import pandas as pd
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

planets = sns.load_dataset("planets")
print(planets.head())
print(planets.isnull().sum())
planets_numeric = planets.select_dtypes(include=['float64', 'int64'])
imputer = IterativeImputer(max_iter=10, random_state=0)

planets_imputed = pd.DataFrame(
    imputer.fit_transform(planets_numeric),
    columns=planets_numeric.columns
)

print("Original planets data")
print(planets_numeric.head(20))
print("\nPlanets data after MICE")
print(planets_imputed.head(20))
```
